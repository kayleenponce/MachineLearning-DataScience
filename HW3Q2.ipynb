Question C: Generate the feature matrix and target vector (target is “balance” in this dataset). 

```python
# Creating the Feature Matrix for Credit dataset:

#creating a phython list of feature names that would like to pick from the dataset:
feature_cols = ['Income','Limit','Rating','Cards','Age','Education','Married']

# use the above list to select the features from the original DataFrame
X =cred_df[feature_cols]

# "Balance" from the data set
y =cred_df['Balance']

# print the Features
print(X.head())
print("\n")

# print the target vector
print(y.head())
```

```python
# normalize the features columns

def feature_scaling(x):
    return x/max(x)
```

```python
# applying function to normaize feature

cred_df['Income']=feature_scaling(cred_df['Income'])
cred_df['Limit']=feature_scaling(cred_df['Limit'])
cred_df['Rating']=feature_scaling(cred_df['Rating'])
cred_df['Cards']=feature_scaling(cred_df['Cards'])
cred_df['Age']=feature_scaling(cred_df['Age'])
cred_df['Education']=feature_scaling(cred_df['Education'])
```

Question D: Split the dataset into testing and training sets with the following parameters: test_size=0.24, random_state=9.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.24,random_state=9)
```

Question E: Use Linear Regression to train a linear model on the training set. Check the coefficients of the linear regression model. Which feature is the most important? Which feature is the least important?

```python
from sklearn.linear_model import LinearRegression

# In the following line, "my_linreg" is instantiated as an "object" of LinearRegression "class". 
my_linreg = LinearRegression()

# fitting the model to the training data:
my_linreg.fit(X_train, y_train)
```

```python
# printing Theta0 using attribute "intercept_":
print(my_linreg.intercept_)

# printing [Theta1, Theta2, Theta3] using attribute "coef_":
print(my_linreg.coef_)

#Most important feature / Least Important Feature
print("Most importannt feature is Cards, Least important is Balance")
```

Question F: Predict “balance” for the users in testing set. Then, compare the predicted balance with the actual balance by calculating and reporting the RMSE.

```python
# make predictions on the testing set
y_prediction = my_linreg.predict(X_test)

print(y_prediction)
```

```python
from sklearn import metrics
import numpy as np

# Calculating "Mean Square Error" (MSE):
mse = metrics.mean_squared_error(y_test, y_prediction)

# Using numpy sqrt function to take the square root and calculate "Root Mean Square Error" (RMSE)
rmse = np.sqrt(mse)

print(rmse)
```

Question G: Use 10-fold Cross-Validation to evaluate the performance of a linear regression in predicting the balance. Thus, rather than splitting the dataset into testing and training, use Cross-Validation to evaluate the regression performance. What is the RMSE when you use cross validation?

```python
# importing the method:
from sklearn.model_selection import cross_val_score
```

```python
# Applying 10-fold cross validation with "logistic regression" classifier:
# In the following line, "my_logreg" is instantiated as an "object" of LogisticRegression "class". 

my_linreg = LinearRegression()

mse_list = cross_val_score(my_linreg, X, y, cv=10, scoring='neg_mean_squared_error')

print(mse_list)
```

```python
# Notice that "cross_val_score" by default provides "negative" values for "mse" to clarify that mse is error.
# in order to calculate root mean square error (rmse), we have to make them positive!
mse_list_positive = -mse_list

# using numpy sqrt function to calculate rmse:
rmse_list = np.sqrt(mse_list_positive)
print(rmse_list)
```

```python
# calculate the average RMSE as final result of cross validation:
print(rmse_list.mean())
```
